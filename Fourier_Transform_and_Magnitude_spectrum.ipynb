```python
import numpy as np
import librosa
import os
import IPython.display as ipd
import matplotlib.pyplot as plt
```

```python
#Loading a c4 on a piano and a c5 of a violin

c4_piano_file= "c4_piano.wav"
c5_violin_file="c5_violin.wav"
```

```python
ipd.Audio(c4_piano_file)
```

```python
ipd.Audio(c5_violin_file)
```

```python
#Loading the files as a wave form

c4_piano,sr=librosa.load(c4_piano_file)
c5_violin,sr=librosa.load(c5_violin_file)
```

```python
c4_piano.shape
```

```python
#Doing the fast fourier transform and then have a look at the first value. This is a complex number!

np.fft.fft(c4_piano)[0]
```

```python
#Defining a function which plots our magnitude spectrum

def plot_magnitude_spectrum(signal,title,sr,f_ratio=1): #We put in our signal, a title, the sample rate and f_ratio, which allows us to represent only smaller frequencies
    ft=np.fft.fft(signal)
    magnitude_spectrum=np.abs(ft) #For the magnitude, we have to take the absolute value of the Fourier Transform
    
    plt.figure(figsize=(20, 5))
    
    frequency=np.linspace(0,sr,len(magnitude_spectrum))
    num_frequency_bins=int(len(frequency)*f_ratio)
 
    
    plt.plot(frequency[:num_frequency_bins],magnitude_spectrum[:num_frequency_bins])
    plt.xlabel("frequency")
    plt.title(title)
    
    plt.show()
    
```

```python
#Plotting magnitude spectrum of c4_piano. Just play with f_ratio. You will see what happens!

plot_magnitude_spectrum(c4_piano,"piano",sr,0.05)

```

```python
plot_magnitude_spectrum(c5_violin,"violin",sr,0.1)
```
